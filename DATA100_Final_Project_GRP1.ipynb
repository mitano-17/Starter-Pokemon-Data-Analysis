## Replacing Nulls and Missing Values


Datasets should be complete to allow quality results and full exploration of the data. To determine which columns contain null values, use the `.isna()` function and then `.sum()` to get the total sum of null values in that column.

```python
df_pokedex_c.isna().sum()
```

We can filter the dataset to only show a list of columns containing null values by using a `.tolist()` method.

```python
df_pokedex_c.columns[df_pokedex_c.isnull().any()].tolist()
```

The columns `weight_kg`, `ability_1`, `growth_rate`, `egg_type_1`, and `egg_cycles` will have to be completed since every Pokemon have these properties. Other columns such as `type_2`, `ability_2`, `ability_hidden` and `egg_type_2` can be taken out of consideration as some Pokemon are not dual-type nor contain more than one abilities and egg type.

For quantitative variables, get the mean using the `.mean()` method and then use `.fillna()` to replace the null values with the computed mean.

```python
weight_mean = df_pokedex_c["weight_kg"].mean()
df_pokedex_c["weight_kg"] = df_pokedex_c["weight_kg"].fillna(weight_mean)

egg_cycles_mean = df_pokedex_c["egg_cycles"].mean()
df_pokedex_c["egg_cycles"] = df_pokedex_c["egg_cycles"].fillna(egg_cycles_mean)

catch_rate_mean = df_pokedex_c["catch_rate"].mean()
df_pokedex_c["catch_rate"] = df_pokedex_c["catch_rate"].fillna(catch_rate_mean)

base_exp_mean = df_pokedex_c["base_experience"].mean()
df_pokedex_c["base_experience"] = df_pokedex_c["base_experience"].fillna(base_exp_mean)

base_frnd_mean = df_pokedex_c["base_friendship"].mean()
df_pokedex_c["base_friendship"] = df_pokedex_c["base_friendship"].fillna(base_frnd_mean)

percent_male_mean = df_pokedex_c["percentage_male"].mean()
df_pokedex_c["percentage_male"] = df_pokedex_c["percentage_male"].fillna(base_frnd_mean)
```

For qualitative variables, get the mode instead of the mean using the `.mode()` method and then use `.fillna()` to replace the null values with the mode.

```python
ability_1_mode = df_pokedex_c["ability_1"].mode()
df_pokedex_c["ability_1"] = df_pokedex_c["ability_1"].fillna(ability_1_mode[0])

egg_type_1_mode = df_pokedex_c["egg_type_1"].mode()
df_pokedex_c["egg_type_1"] = df_pokedex_c["egg_type_1"].fillna(egg_type_1_mode[0])

growth_rate_mode = df_pokedex_c["growth_rate"].mode()
df_pokedex_c["growth_rate"] = df_pokedex_c["growth_rate"].fillna(growth_rate_mode[0])
```

Now, we can check the columns again and identify whether the columns we modified above no longer have null values.

```python
df_pokedex_c.columns[df_pokedex_c.isnull().any()].tolist()
```

## Normalization

<b>Normalization</b> is a method of balancing our data by lessening the impact of outliers in the dataset. To prove that our data is not normalized, make a Q-Q plot using the `.probplot()` function.

```python
stats.probplot(df_pokedex_c["attack"], dist="norm", plot=plt)
```

As we can see above, the line that the data created is not a straight line. Therefore, we can convince ourselves that the data is not normalized. To normalize our data, use the `.boxcox()` method from the stats library. The <b>box-cox transformation</b> makes the variance consistent across all the data points.

```python
normal_attack, normal_lambda = stats.boxcox(df_pokedex_c["attack"])
```

We can plot again a Q-Q plot, this time using the normalized data that we computed using box-cox transformation.

```python
stats.probplot(normal_attack, dist="norm", plot=plt)
```

Looking at the plot above, we notice that the data points create a straighter line compared to when it was not normalized. Therefore, we can repeat the process to the rest of the statistics and then replace the old data with the new one.

```python
normal_hp, normal_lambda = stats.boxcox(df_pokedex_c["hp"])
normal_defense, normal_lambda = stats.boxcox(df_pokedex_c["defense"])
normal_sp_attack, normal_lambda = stats.boxcox(df_pokedex_c["sp_attack"])
normal_sp_defense, normal_lambda = stats.boxcox(df_pokedex_c["sp_defense"])
normal_speed, normal_lambda = stats.boxcox(df_pokedex_c["speed"])
```

```python
df_pokedex_c["hp"] = normal_hp
df_pokedex_c["attack"] = normal_attack
df_pokedex_c["defense"] = normal_defense
df_pokedex_c["sp_attack"] = normal_sp_attack
df_pokedex_c["sp_defense"] = normal_sp_defense
df_pokedex_c["speed"] = normal_speed
```

```python
df_pokedex_c[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]]
```

## Feature Scaling

Clustering requires data that is standardized - meaning the data have a mean of 0 and standard deviation of 1. <b>Standardizing</b> our data helps the algorithm since it repositions data points to allow for faster computation and minimal use of resources.

To standardize our data, use the `StandardScaler()` function to transform our data. This function comes from the sklearn library best used for machine learning.

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardize = StandardScaler()
standard_stats = standardize.fit_transform(df_pokedex_c[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]])
df_pokedex_c[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]] = standard_stats
```

Once done with both normalization and standardization, we can proceed with clustering our data.

## Elbow Method

The K-Means clustering algorithm requires how many clusters we can use to group our data. Intuitively, we can set the number to 18 based on the elemental types Pokemon have. However, the algorithm detects patterns and groups Pokemon based on the pattern. To determine the number of clusters (or Ks) we have to use, we can utilize the elbow method.

```python
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
```

```python
inertia = []

for x in range(1, 20):
    clusters = KMeans(n_clusters = x)
    clusters.fit(standard_stats)
    inertia.append(clusters.inertia_)
```

After that, use the `KneeLocator().elbow` function from the kneed library. This can determine the elbow/knee of the data; that is, the number of clusters the algorithm should use. The elbow can also be visualized using the `.plot()` function. Based on the KneeLocator, the optimal amount of clusters is <b>4</b>. The plot supports this since the inertia starts decreasing in a stable manner once the x number of clusters reaches 4.

```python
from kneed import KneeLocator
```

```python
cluster_n = KneeLocator(range(1, 20), inertia, curve="convex", direction="decreasing").elbow
```

```python
cluster_n
```

```python
plt.plot(range(1, 20), inertia)

plt.title("Elbow Method")
plt.xlabel("X amount of clusters")
plt.ylabel("Y inertia")

plt.show()
```

## K-Means Clustering

Now that we have preprocessed our data and determined the right amount of clusters to use, we can now implement the clustering algorithm using the `KMeans()` function and then use the `.fit()` function to compute for the mean and standard deviation of the data. Lastly, get the labels of each cluster using the `.labels_` method.

```python
df_pokedex_c["cluster"] = KMeans(cluster_n, init = 'random', n_init = 10, random_state = 1).fit(standard_stats).labels_
```

# ðŸ“ Hypothesis Testing ðŸ“

To verify if there really is a difference in all clusters, we can conduct <b>hypothesis testing</b>. To perform hypothesis testing, let us propose our null and alternative hypothesis.
- <b>H<sub>O</sub>:</b> There is no difference between clusters.
- <b>H<sub>A</sub>:</b> There is a difference between clusters.

Define the significance level at <b>Î± = 0.05</b>. The significance level serves as the bar to determine if there is a significant difference between the significance level and p-value. The <b>p-value</b> is derived from calculating the Z-score.

To compare two or more independent groups, we use the <b>one-way ANOVA</b> test. Use the `.stats.f_oneway()` to get the p-value.

```python
significance_level = 0.05
```

```python
cluster_0 = df_pokedex_c[df_pokedex_c["cluster"] == 0]
cluster_0 = cluster_0[["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "cluster"]].reset_index()

cluster_1 = df_pokedex_c[df_pokedex_c["cluster"] == 1]
cluster_1 = cluster_1[["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "cluster"]].reset_index()

cluster_2 = df_pokedex_c[df_pokedex_c["cluster"] == 2]
cluster_2 = cluster_2[["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "cluster"]].reset_index()

cluster_3 = df_pokedex_c[df_pokedex_c["cluster"] == 3]
cluster_3 = cluster_3[["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "cluster"]].reset_index()
```

```python
Fscore, Pval = stats.f_oneway(cluster_0[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]],
                              cluster_1[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]],
                              cluster_2[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]],
                              cluster_3[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]])
```

```python
print("F-Score: ", Fscore, "\nP-Value: ", Pval)
```

The <b>F-score</b> is the rate of how precise the data was modeled while the <b>p-value</b> is the level of how significant the data has been revealed. To determine if we use the null or alternate hypothesis for verification, we compare the p-values with the significance level.

```python
if Pval[0] < significance_level:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

```python
if Pval[1] < significance_level:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

```python
if Pval[2] < significance_level:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

```python
if Pval[3] < significance_level:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

Due to the pvalue being less than the significance level, we reject the null hypothesis. Thus, we can conclude that <b>there is a difference between the clusters</b>.
